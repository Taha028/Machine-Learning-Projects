# Analyzing the Salary Column and Concluding Which Central Tendency Will Be Use

```python
#As you can see the data is right skewed so we dont use mean as ,
#it is highlt effected by outlier so we use median here as its a best practice to,
#fill values with median and median is also not effected by outliers.

fig,ax = plt.subplots(figsize = (8,8))
sns.distplot(data["salary"])
```

```python
#There is a very simple method to fill values 
data["salary"].fillna(data["salary"].median(),inplace = True)
```

```python
#As you can see values are filed by median. if you do it with mean and mode same code just change central tedency.
data.isnull().sum()
```

# Now imputing missing values

```python
#Imputing Values by median
for i in unique_val:
    strr = []
    for j,k in data.iterrows():
        if k["specialisation"] == i:
            strr.append(k["salary"])
    med = np.median(strr)
    data.loc[data["specialisation"]==i,"salary"] = data.loc[data["specialisation"]==i,"salary"].fillna(med)
    
            
```

```python
#If you have to calculate mean group wise
for i in unique_val:
    count = 0
    sal = 0
    for j,k in data.iterrows():
        if k["specialisation"] == i:
            if not pd.isnull(k["salary"]):
                sal+=k["salary"]
                count+=1
    if count>0:
        p = sal/count
        data.loc[data["specialisation"]==i,"salary"] = data.loc[data["specialisation"]==i,"salary"].fillna(sal)
        
```
